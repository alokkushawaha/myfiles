```python
#Geopy has been used for the latitude and longitude extraction for supply locations 

from geopy.exc import GeocoderTimedOut
from geopy.geocoders import Nominatim
longitude = []
latitude = []
def findGeocode(city):
    try:
        geolocator = Nominatim(user_agent="ALOK kumar")
        return geolocator.geocode(city)
    except GeocoderTimedOut:
          
        return findGeocode(city)  
for i in (df2["Source"]):
    if findGeocode(i) != None:
        loc = findGeocode(i)
        latitude.append(loc.latitude)
        longitude.append(loc.longitude)
    else:
        latitude.append(np.nan)
        longitude.append(np.nan)
```

```python
#Geopy has been used for the distance calculation between two locations 

from  geopy.distance import geodesic
coordinate=[]

source = (float(df2.iloc[0][1]),float(df2.iloc[0][2]))
          
for i in range(len(df)):
    a=(df.iloc[i][2]),float(df.iloc[i][3])
    coordinate.append(a)
```

```python
dist=[] # Distance matrix has been obtained from the latitude and longitude
for i in range(len(coordinate)):
    d=geodesic(source, coordinate[i]).km
    dist.append(d)
    
```

```python
dist
```

```python
import gurobipy as gp
from gurobipy import GRB
from itertools import product
```

```python
demand_point=[i for i in range(len(df))] #Number of demand point
quality=[i for i in range(3)] #Number of the qualities
distance=[dist,
          dist,
          dist] #Here, all qualities need to be distributed at same distance so three distance rows taken for three qualities
supply=[1500,800,1000] # Supply for the three quality of the mango 

demand=[] # Demand matrix . It has been generated from the data which can be given as input from the data frame defined for the city and demand
for i in range(len(df)):
    b=df.iloc[i][1]
    demand.append(b)
    
time=[48,36,24] # Time for the ith quality mango degradation obtained values by the simulation

v=10 #Speed for the vehicle used for the transportation

unit_distance_cost=10 #Per km transportation cost
```

```python
#Model defination for the GurobiPy

m=gp.Model('quality_model')
```

```python
#Gurobipy variable declaration

x=m.addVars(list(product(quality,demand_point)),vtype=GRB.CONTINUOUS) #Amount to be transported
y=m.addVars(list(product(quality,demand_point)),vtype=GRB.BINARY) #binary variable
z=m.addVars(list(product(quality,demand_point)),vtype=GRB.CONTINUOUS) #Difference of time for degradation and time for transportation
```

```python
x
```

```python
y
```

```python
m.addConstrs((gp.quicksum(x[(i,j)] for i in quality) )>=demand[j] for j in demand_point) #Demand constraint
```

```python
m.addConstrs((gp.quicksum(x[(i,j)] for j in demand_point )<=supply[i] for i in quality)) #Supply constraint
```

```python
m.addConstrs((z[i,j])==(time[i]-distance[i][j]/v) for i in quality for j in demand_point) # for the quality degradation
```

```python
#Indicator constraint used to model if else conditional statements. if zij<=0 then yij=0, if zij>=0 yij==1

m.addConstrs((y[i,j]==0)>>(z[i,j]<=0)  for i in quality for j in demand_point)
m.addConstrs((y[i,j]==1)>>(z[i,j]>=0)  for i in quality for j in demand_point)
```

```python
#xij<=M*yij

m.addConstrs(x[(i,j)] <=1000000*y[(i,j)] for i in quality for j in demand_point)
```

```python
#Objective function minimize the net cost
m.setObjective(gp.quicksum(x[(i,j)] * distance[i][j]*unit_distance_cost for i in quality for j in demand_point),GRB.MINIMIZE)
```

```python
#Optimization statements
m.optimize()
```
